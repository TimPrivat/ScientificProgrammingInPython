<h3>Task 2</h3>
<p>In Task 2 we implement python functions to determine the corresponding lossvalues between the predicted and the actual data. We implement the same function twice. One function uses pythons builtin list to store and return the values and the other one uses numpy arrays.</p>

<h4>Loss Function</h4>
<p>The formula for the loss function is defined as follows[5]: </p>
$$ Loss=\alpha *\left| R^{Pred.}-R^{Exp.}\right| + \beta * \left| E^{Pred.}-E^{Exp.}\right| $$ 

<h4>Implementation</h4>
<p>First we implement a function that internally uses pythons list to store the data:</p>

```python
def LossAsList(alpha,beta,expected,prediction):
    """ 
    This function determines the lossfunction between predicted
    and recoreded weatherdata.
    It uses a list to store and return the calculated data.
    
    :param alpha: rainfall weighing factor
    :param beta: evaporation weighing factor
    :param expected: a pandas dataframe holding the actual recorded data
    :param prediction: a panadas dataframe holding the predicted data

    :return: a python list where every index is the loss of the corresponding 
             indices between the predicted and actual data. 
                
    """
    if not isinstance(alpha, float):
        raise TypeError('alpha must be provided as a float')
    elif not isinstance(beta, float):
        raise TypeError('beta must be provided as a float')
    elif not isinstance(expected, pd.DataFrame):
        raise TypeError('expected must be provided as a pandas dataframe')
    elif not isinstance(prediction, pd.DataFrame):
        raise TypeError('prediction must be provided as a pandas dataframe')
    
    expectedrain = expected["Rainfall"].tolist()
    expectedevaporation = expected["Evaporation"].tolist()

    predictedrain = prediction["Rainfall Pred."].tolist()
    predictedevaporation = prediction["Evaporation Pred."].tolist()

    loss = []

    for(pred_rain, exp_rain, pred_eva, exp_eva) in \
        zip(predictedrain, expectedrain, predictedevaporation, expectedevaporation):
        
        currentloss = alpha * abs(pred_rain - exp_rain) + beta * abs(pred_eva - exp_eva)
        loss.append(currentloss)
    return loss
```

<p>After that we implemet the same function again, but now it uses a numpy array to store the data:</p>

```python
def LossAsNP(alpha,beta,expected,prediction):
    """ 
    This function determines the lossfunction between predicted
    and recoreded weatherdata.
    It uses a numpy array to store and return the calculated data.
    
    :param alpha: rainfall weighing factor
    :param beta: evaporation weighing factor
    :param expected: a pandas dataframe holding the actual recorded data
    :param prediction: a panadas dataframe holding the predicted data

    :return: a numpy array where every index is the loss of the corresponding 
             indices between the predicted and actual data. 
                
    """
    if not isinstance(alpha, float):
        raise TypeError('alpha must be provided as a float')
    elif not isinstance(beta, float):
        raise TypeError('beta must be provided as a float')
    elif not isinstance(expected, pd.DataFrame):
        raise TypeError('expected must be provided as a pandas dataframe')
    elif not isinstance(prediction, pd.DataFrame):
        raise TypeError('prediction must be provided as a pandas dataframe')
        
    expectedrain = expected["Rainfall"].to_numpy()
    expectedevaporation = expected["Evaporation"].to_numpy()

    predictedrain = prediction["Rainfall Pred."].to_numpy()
    predictedevaporation = prediction["Evaporation Pred."].to_numpy()
    
    loss = alpha * np.abs(predictedrain - expectedrain) + \
           beta * np.abs(predictedevaporation - expectedevaporation)
    return loss    
```

<h3>Task 3</h3>
<p>In task 3 we run both of our functions and time how long they take for 100 iterations. Then we compare which implementation is faster: The python list or the numpy array.</p>

<h4>Constants</h4>
<p>For this we need to assign values to the weighing factors α and β. These are project specific and were given to us in the assignement paper.<br>
We define them as follows:
$$\alpha =0.5$$
$$\beta =0.5$$
</p>

```python
alpha = 0.5
beta = 0.5
```

<h4>Computing our data</h4>
<p>Now that we have all the data we need, we can use our functions to determine the speed difference between both implementations.<br>
For this we use pythons builtin module timeit. We run 100 iterations of our functions and print the resulting time it took to compute:</p>

```python
ListTime = timeit.timeit(lambda:LossAsList(alpha,beta,weather_experiment,weather_prediction), number=100)
NPArrayTime = timeit.timeit(lambda:LossAsNP(alpha,beta,weather_experiment,weather_prediction), number=100)

print(f"The List implementation finished after {ListTime} seconds.")
print(f"The Numpy implementation finished after {NPArrayTime} seconds.")
```

```python
times=[ListTime, NPArrayTime]
mintime = min(times)
maxtime = max(times)

difference = maxtime/mintime

if mintime==ListTime:
    print(f"The List implementation is {difference} times faster than the Numpy implementation.")
else:
    print(f"The Numpy implementation is {difference} times faster than the List implementation.")
```
